```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from pathlib import Path

print ("libraries loaded successfully")
```

```python
data_dir = Path("../data")

df = pd.read_csv(data_dir / "still_01.csv")

print(df.shape)
df.head()
```

```python
time = df.iloc[:, 0]
acc_x = df.iloc[:, 1]
acc_y = df.iloc[:, 2]
acc_z = df.iloc[:, 3]

plt.figure(figsize=(12, 4))
plt.plot(time, acc_x, label="X", alpha=0.7)
plt.plot(time, acc_y, label="Y", alpha=0.7)
plt.plot(time, acc_z, label="Z", alpha=0.7)
plt.xlabel("Time (s)")
plt.ylabel("Acceleration (m/s^2)")
plt.title("Still")
plt.legend()
plt.grid(True, alpha=0.3)
plt.show()
```

```python
def plot_acceleration(csv_filename, title):
    df = pd.read_csv(data_dir / csv_filename)

    time = df.iloc[:,0]
    acc_x = df.iloc[:,1]
    acc_y = df.iloc[:,2]
    acc_z = df.iloc[:,3]

    plt.figure(figsize=(12,3))
    plt.plot(time, acc_x, label="X")
    plt.plot(time, acc_y, label="Y")
    plt.plot(time, acc_z, label="Z")
    plt.title(title)
    plt.xlabel("time(s)")
    plt.ylabel("acceleration(m/s^2)")
    plt.legend()
    plt.grid(True, alpha=0.4)
    plt.show()
```

```python
plot_acceleration("still_01.csv", "Still")
plot_acceleration("walk_01.csv", "Walk")
plot_acceleration("run_01.csv", "Run")
```

```python
def plot_acceleration_with_trimming(csv_filename, title):
    df = pd.read_csv(data_dir / csv_filename)

    time_col = df.iloc[:,0]
    mask = (time_col >= 3) & (time_col <= time_col.max() - 3)
    df = df[mask].reset_index(drop = True)

    time = df.iloc[:,0]
    acc_x = df.iloc[:,1]
    acc_y = df.iloc[:,2]
    acc_z = df.iloc[:,3]

    plt.figure(figsize=(12,3))
    plt.plot(time, acc_x, label="X")
    plt.plot(time, acc_y, label="Y")
    plt.plot(time, acc_z, label="Z")
    plt.title(title)
    plt.xlabel("time(s)")
    plt.ylabel("acceleration(m/s^2)")
    plt.legend()
    plt.grid(True, alpha=0.4)
    plt.show()
```

```python
plot_acceleration("still_01.csv", "Still")
plot_acceleration_with_trimming("walk_01.csv", "Walk")
plot_acceleration_with_trimming('run_01.csv', "Run")
```
